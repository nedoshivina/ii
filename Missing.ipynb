```python
import pandas as pd 
import numpy as np

#Создание дата сета
data = [['Вася', np.nan, 70, 4, 7], ['Петя', 330, 50, np.nan, 12], ['Аня', 200, 70, 3.2, 1], 
        ['Коля', np.nan, 80, 5, np.nan], ['Оля', 260, np.nan, 4.2, 8]]
df = pd.DataFrame(data, columns=['Name', 'attendance', 'counter_study', 'GPA', 'counter_rep']) 
df
```

```python
#Вычисление пропущенных значений в каждой колонке
percent_missing = df.isnull().mean() * 100
percent_missing 
```

```python
#Отбрасывание строк, где есть NaN
df_no_missing = df.dropna()
df_no_missing
```

```python
#Отбрасывания строки с большим количеством пропусков
df['num_missing'] = df.isnull().sum(axis = 1)

# Определяем индексы строк, в которых количество пропущенных значений > 1
ind_missing = df[df['num_missing'] > 1].index

# Удаляем строки с большим количеством пропусков (0 - отбрасываем строку, 1 - столбец)
df_less_missing_rows = df.drop(ind_missing, axis=0)

df_less_missing_rows = df_less_missing_rows.drop('num_missing', axis=1)
df = df.drop('num_missing', axis=1)
df_less_missing_rows
```

```python
#Отбрасывание колонки с большим % пропусков
cols_to_drop = ['attendance']

data_less_attendance = df.drop(cols_to_drop, axis = 1)
data_less_attendance
```

```python
#Заполнение пропусков нулем, медианой или средним значением
median = df['attendance'].median()
mean = df['GPA'].mean()

df['counter_rep'] = df['counter_rep'].fillna(0)
df['attendance'] = df['attendance'].fillna(median)
df['GPA'] = df['GPA'].fillna(mean)
df
```
